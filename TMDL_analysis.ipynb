### Quick look at what we got

```python
import collections
plot_cols = ['Region', 'Lead State', 'Document Type', 'TMDL Status',
       'Fiscal Year', 'TMDL Type', 'Pollutant', 'Pollutant Group']

for i, col in enumerate(plot_cols): 
    cnt = df[col].value_counts()
    cnt = collections.OrderedDict(sorted(cnt.items(), key=lambda t: t[0]))
    plt.figure(num=i, figsize=(18, len(cnt)*.4))
    ax = plt.subplot(1, 1,1)
    #print(list(cnt.keys()))
    ax.barh(range(0,len(cnt)), list(cnt.values()))
    #ax.set_yscale('log')
    plt.ylabel(col)
    plt.yticks( range(0,len(cnt)),list(cnt.keys()),va='bottom')
```
