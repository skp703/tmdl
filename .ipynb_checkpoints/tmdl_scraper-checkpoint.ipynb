```python
def display(l, t = 8):
    length = len(l)
    if length < t:
        return (", ".join(map(str,l)))
    else:
        return (", ".join(map(str,l[:t-1])) + " and +{} others".format(length - (t - 1)))
print("What do we have..........")
for col in df.columns:
    uq = df[col].unique()
    print("{0:30}{1:10}: {2:100}".format(col,len(uq),display(uq)))
```

```python
# there are 136 document types, essentially those can be divided in 4-- 
# TMDL Report, TMDL Approval, some mixture of Approval/Report and unknowns
# I will try to find reports and limit my analysis to reports or Documents
uq_dt = pd.DataFrame(df["Document Type"].unique())  # unique doc types
uq_dt[1]= uq_dt[0]  # just a copy.... quick and dirty
uq_dt = uq_dt.replace({1:{"TMDL Document":"Document",  # all these are essentially same category so replace col 1 to document
                          "TMDL_DOCUMENT":"Document", 
                          "TMDL Report":"Document",
                          "TMDL Report and Appendices":"Document",
                          "Executive Summary/TMDL Report/Appendices/Response to Comments":"Document",
                          'TMDL Report/Appendices':"Document",
                          'TMDL Report/Appendices(Revised Version with Errata)':"Document",
                         }})

# anything that starts with "TMDL Document -" is also a document
uq_dt[1] = uq_dt[1].map(lambda val : "Document" if str(val).startswith("TMDL Document -") else val)
#print(uq_dt[1].unique())
import re
pat = re.compile("^(?!.*eport.*).*?(etter|ational).*?")  # has letter or Rationale but no report make those 'Letter'
uq_dt[1] = uq_dt[1].map(lambda val : "Letter" if pat.match(str(val)) else val)
pat = re.compile(".*?eport.*?")  # any thing still with report is probabaly a doc/letter
uq_dt[1] = uq_dt[1].map(lambda val : "Doc/Letter" if pat.match(str(val)) else val)
#print(uq_dt[1].unique())
uq_dict = uq_dt[uq_dt[1].isin(['Document','Letter','Doc/Letter']) ]
d_rep = uq_dict.set_index(0).to_dict()
#print(d_rep[1])
df= df.replace({"Document Type":d_rep[1]})  # convert document type to one of the three Document','Letter','Doc/Letter'

# covert APPROVED/ESTABLISHED to Approved/Established
df= df.replace({"TMDL Status":{"APPROVED/ESTABLISHED":"Approved/Established"}})
```

```python
# lets do tte actual grouping by report link discussed earlier
# the aggregation function (how to combine multiple rows.. is simply join unique vales by ",")
def join_strings(s):
   return ";".join([str(y) for y in np.unique(s)])

df_grouped = df.groupby("Report Link").agg({'Region' : "first", 
                                           'Lead State': join_strings,
                                           'TMDL ID': join_strings,
                                           'TMDL Name': join_strings,
                                           'Document Name': join_strings,
                                           'Document Type': join_strings,
                                           'TMDL Status': join_strings,
                                           'Fiscal Year': np.max, # multi year??
                                           'TMDL Type': join_strings,
                                           'Pollutant': join_strings,
                                           'Pollutant Group': join_strings,
                                           'Assessment Unit ID': join_strings,
                                           'MatchScoreDescending':np.sum, # interesting; since all are 100 the sum will give join count
                                           'Ammonia':np.sum,
                                           'Chlorine':np.sum,
                                           'Dioxins':np.sum, 
                                           'Habitat Alterations':np.sum,
                                           'Mercury':np.sum,
                                           'Metals (other than Mercury)':np.sum,
                                           'Noxious Aquatic Plants':np.sum,
                                           'Nutrients':np.sum,
                                           'Oil and Grease':np.sum,
                                           'Organic Enrichment/Oxygen Depletion':np.sum, 
                                           'Other Cause':np.sum,
                                           'Pathogens':np.sum, 
                                           'Pesticides':np.sum, 
                                           'Polychlorinated Biphenyls (PCBs)':np.sum, 
                                           'Salinity/Total Dissolved Solids/Chlorides/Sulfates':np.sum, 
                                           'Sediment':np.sum,
                                           'Temperature':np.sum, 
                                           'Total Toxics':np.sum, 
                                           'Toxic Inorganics':np.sum, 
                                           'Toxic Organics':np.sum,
                                           'Trash':np.sum, 
                                           'Turbidity':np.sum, 
                                           'pH/Acidity/Caustic Conditions':np.sum}
                                          )
```

```python
# download all files and extract text
# relies on xpdf (pdftotext.exe in the folder)
import os
import urllib
import subprocess
path = r"C:\\tmdl_rep";
def download_file(download_url, file_name,path):
    if not os.path.exists(os.path.join(path,file_name)):
        print("downloading url:{} in file: {}".format(row[0],row[1]["Document Name"]))
        with urllib.request.urlopen(download_url) as web_file, open(os.path.join(path,file_name),'wb') as local_file:
            local_file.write(web_file.read())
            local_file.close()
    if not os.path.exists(os.path.join(path,file_name)+".txt"):
        print("extracting text for",os.path.join(path,file_name))
        subprocess.call([os.path.join(path,"pdftotext.exe"),os.path.join(path,file_name),os.path.join(path,file_name)+".txt"])
    if not os.path.exists(os.path.join(path,file_name)+".txt"):  #conversion failed remove copy protection
         subprocess.call([os.path.join(path,"qpdf.exe"),"-decrypt",os.path.join(path,file_name),os.path.join(path,file_name)+".pdf"])
         subprocess.call([os.path.join(path,"pdftotext.exe"),os.path.join(path,file_name)+".pdf",os.path.join(path,file_name)+".txt"])
        
        
for row in df_grouped.iterrows():
    download_file(row[0], row[1]["Document Name"],path)
    
```

```python

pdfreader = "SumatraPDF.exe"
def manually_verify(df_verify):  # manually verify reports, returns true if report
    return_vals=[]
    count = 0
    for row in df_verify.iterrows():
        file_name = row[1]["Document Name"]
        count =+1
        print(count,file_name)
        p = subprocess.Popen([os.path.join(path,pdfreader),os.path.join(path,file_name)])  # open pdf
        nb = input('if NOT a report press N and return, just press return if a report')
        print("Report",nb!="n")
        p.kill()
        return_vals.append(nb!="n")
    return return_vals
```

```python
# rule out some obvious issues with reports

# reports with - copy are duplicates
print(df_grouped['Document Name'][(df_grouped['Document Name'].str.contains(".*(?i)Copy.pdf"))])
df_grouped = df_grouped[np.invert(df_grouped['Document Name'].str.contains(".*(?i)Copy.pdf"))] 

# lets look for reports with some version of decision doc in them 
regex = "(?i).*(decision|dec.{0,3}doc|approv).*"
print(df_grouped['Document Name'][(df_grouped['Document Name'].str.contains(regex))].shape)
# manually verify these 
remove = manually_verify(df_grouped[(df_grouped['Document Name'].str.contains(regex))])
if sum(remove) > 0: 
    df_grouped = df_grouped.drop(df_grouped[(df_grouped['Document Name'].str.contains(regex))][np.invert(remove)].index)
    
# lets look for appendix with app in them 
regex = "(?i).*(app).*(?<!approv(ed|al))" # negative look-behind assertion to verify that the text just matched does not equal approv
print(df_grouped['Document Name'][(df_grouped['Document Name'].str.contains(regex))].shape)
# manually verify these 
remove = manually_verify(df_grouped[(df_grouped['Document Name'].str.contains(regex))])
if sum(remove) > 0: 
    df_grouped = df_grouped.drop(df_grouped[(df_grouped['Document Name'].str.contains(regex))][np.invert(remove)].index)

# lets look at doc with "decision document" in the text 
pat = re.compile("(?i)(\s)decision(\s|-)document(\s)")
watch = []
for row in df_grouped.iterrows():
    file_name = row[1]["Document Name"]+".txt"  # row[1]["Document name"]
    file_path = os.path.join(path,file_name)
    with open (file_path, "r") as f:
        data=f.read()
        match =pat.finditer(data)
        matches = list(match)
        if len(matches) > 0:
            print(row[0], len(matches),",".join([m.group() for m in matches]))
            watch.append(row[0])

remove = manually_verify(df_grouped.ix[watch])
if sum(remove) > 0: 
    df_grouped = df_grouped.drop(df_grouped.ix[watch][np.invert(remove)].index)                
```

```python
import re
re_test_words = dict()
df_grouped["likely_model"] = ""
df_grouped["score1"] = ""
for key in test_words.keys():
    re_test_words[key]=re.compile(test_words[key],)
    df_grouped[key]=0
    df_grouped[key+"match"]=""
```

```python
import collections
path = r"C:\\tmdl_rep"
for row in df_grouped.iterrows():
    file_name = row[1]["Document Name"]+".txt"  # row[1]["Document name"]
    file_path = os.path.join(path,file_name)
    print("******* ",file_name)
    d_model = dict()
    with open (file_path, "r") as f:
        data=f.read()
        for key in re_test_words.keys():           
            match = re_test_words[key].finditer(data)
            matches = list(match)
            d_model[key]=len(matches)
            if len(matches) > 0:
                #print("--------------------",key)
                #print(",".join([m.group() for m in matches]))
                df_grouped.ix[row[0],key] = len(matches)
                df_grouped.ix[row[0],key+"match"] = ",".join([m.group() for m in matches])
    od = collections.OrderedDict(sorted(d_model.items(),key=lambda k: d_model[k[0]],reverse=True))
    score = ""
    models = ""
    for k, v in od.items(): 
        if v>1:
            score = score+";"+str(v)
            models = models+";"+k
    df_grouped.ix[row[0],"likely_model"], df_grouped.ix[row[0],"score1"]=models,score
```
